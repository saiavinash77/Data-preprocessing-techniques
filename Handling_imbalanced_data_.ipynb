# TARGET : making the no .of .rows of legit cases  == no. of .rows of fradulent cases

# separating the legit and fraudulent transactions

```python
credit_card_data.shape
```

```python
legit_cases = credit_card_data[credit_card_data.Class==0]
fradulent_cases= credit_card_data[credit_card_data.Class ==1]
```

```python
legit_cases.shape
```

```python
fradulent_cases.shape
```

# Now Building a sample dataset containing similar distribution of Legit & Fraudulent Transactions

```python
legit_sample_cases =legit_cases.sample(n=25)
```

```python
legit_sample_cases.shape
```

# **now you can see legit cases we had taken only 25 on the above  **

```python
new_dataset = pd.concat([legit_sample_cases,fradulent_cases],axis=0)
```
