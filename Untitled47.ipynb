# **LABEL**-**ENCODING**

*labelencoding* is nothing but converting the values to numerical form which are in the form of alphabetical .
we will do this because ml algo's require numerical input to perform computations.

```python
# Step 1 : importing all the requirements
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

**Now we are working with the  breast cancer data**

```python
cancer_data = pd.read_csv("/content/data.csv")
```

```python
# first 5 rows of the dataframe will be displayead by using this Function
cancer_data.head()
```

```python
#counting how many values different values in our desired coulmn.
cancer_data['diagnosis'].value_counts()
```

```python
#Now its time to load the label encoder function which is here

label_encoder = LabelEncoder()
```

```python
labels = label_encoder.fit_transform(cancer_data.diagnosis)
```

```python
cancer_data['target']=labels
```

```python
cancer_data.head()
```

```python
cancer_data['target'].value_counts()
```
